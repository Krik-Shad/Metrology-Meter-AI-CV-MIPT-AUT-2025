```python
#Подключение нужных библиотек
from os import rename, makedirs, listdir, remove
from cv2 import VideoCapture, imwrite
# import numpy as np
# import pandas as pd
```

```python
#Получение данных видео
video_path = "video_2025-10-31_18-44-56.mp4"

video = VideoCapture(video_path)
makedirs(f"frames", exist_ok=True)
```

```python
#Обработка видео, выделение кадров
index = 0

while True:
    not_ended, frame = video.read()

    if not not_ended:
        break

    if (index%30==0):
        imwrite(f"frames/frame_{index//30}.jpg", frame)
    index+=1

video.release()
```

```python
#Расфасовка по каталогам
classes = [str(50+20*i) for i in range(11)]
times = [
    (3, 15),
    (18, 36),
    (38, 53),
    (58, 77),
    (80, 98),
    (102, 123),
    (126, 163),
    (166, 192), 
    (196, 224),
    (229, 261),
    (264, 299)]
class_index = 0

for class_name in classes:
    makedirs(f"frames/{class_name}", exist_ok=True)

for  class_index, (start, end) in enumerate(times):
    class_name = classes[class_index]
    range_length = end - start + 1
    
    print(f"Обрабатываем класс {class_name}: кадры {start}-{end} (длина: {range_length})")
    
    # Определяем шаг для длинных участков
    step = 2 if range_length > 30 else 1
    
    moved_count = 0
    for i in range(start, end + 1, step):
        try:
            rename(
                f"frames/frame_{i}.jpg",
                f"frames/{class_name}/frame_{i}.jpg"
            )
            moved_count += 1
            print(f"  Кадр {i} -> {class_name}")
        except FileNotFoundError:
            print(f"  Файл frame_{i}.jpg не найден")
    
    print(f"Для класса {class_name} перемещено {moved_count} кадров\n")
```

```python
#удаление ненужных кадров
rest_frames = [f for f in listdir('frames') if f.startswith('frame_') and f.endswith('.jpg')]

deleted_count = 0
for frame_file in rest_frames:
        frame_path = "frames/"+str(frame_file)
        remove(frame_path)
        deleted_count+=1
print(f"Удалено неиспользованных кадров: {deleted_count}")
```
